```python
import datetime
import xlrd
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Get data at:
* Gas/Diesel: http://www.eia.gov/petroleum/gasdiesel/
* Production: http://tonto.eia.gov/dnav/pet/hist/LeafHandler.ashx?n=PET&s=WCRFPUS2&f=W

```python
def csv_from_excel(efile, cfile):
    wb = xlrd.open_workbook(efile)
    csv_file = open(cfile, 'wb')
    wr = csv.writer(csv_file, quoting=csv.QUOTE_ALL)
    snames = wb.sheet_names()
    for s in snames:
        if s == 'Data 1':
            sh = wb.sheet_by_name(s)
            for rownum in xrange(sh.nrows):
                wr.writerow(sh.row_values(rownum))
    csv_file.close()
```

```python
csv_from_excel('WCRFPUS2w.xls', 'prod.csv')
csv_from_excel('pswrgvwall.xls', 'gas.csv')
csv_from_excel('psw18vwall.xls', 'diesel.csv')
```

```python
def csv_to_df(fcsv):
    lines = list(csv.reader(open(fcsv)))
    header, values = lines[2], lines[3:]
    data = {h:v for h,v in zip (header, zip(*values))}
    return pd.DataFrame.from_dict(data)
```

```python
gprod = csv_to_df('prod.csv')
gas = csv_to_df('gas.csv')
diesel = csv_to_df('diesel.csv')
```

```python
cols = gas.columns
cols = [c.replace('Retail Gasoline Prices  (Dollars per Gallon)', '') for c in cols]
cols = [c.replace('Regular Conventional', '') for c in cols]
cols = [c.replace('Weekly', '') for c in cols]
gas.columns = [c.lower().replace(' ', '').replace('.', '') for c in cols]

cols = diesel.columns
cols = [c.replace('Retail Prices  (Dollars per Gallon)', '') for c in cols]
cols = [c.replace('No 2 Diesel', '') for c in cols]
cols = [c.replace('Weekly', '') for c in cols]
diesel.columns = [c.lower().replace(' ', '').replace('.', '') for c in cols]

cols = gprod.columns
cols = [c.replace('Retail Gasoline Prices  (Dollars per Gallon)', '') for c in cols]
cols = [c.replace('Regular Conventional', '') for c in cols]
cols = [c.replace('Weekly', '') for c in cols]
gprod.columns = [c.lower().replace(' ', '').replace('.', '').replace('usfieldproductionofcrudeoil(thousandbarrelsperday)', 'oil') for c in cols]
```

```python
def xldate_to_datetime(xldate):
    temp = datetime.datetime(1900, 1, 1)
    delta = datetime.timedelta(days=xldate)
    return temp+delta
```

```python
gas.date = gas.date.astype(float).apply(xldate_to_datetime)
diesel.date = diesel.date.astype(float).apply(xldate_to_datetime)
gprod.date = gprod.date.astype(float).apply(xldate_to_datetime)
```

```python
gas = gas.replace(r'\s+', np.nan, regex=True).replace(r'', 0.0, regex=True)
diesel = diesel.replace(r'\s+', np.nan, regex=True).replace(r'', 0.0, regex=True)
gprod = gprod.replace(r'\s+', np.nan, regex=True).replace(r'', 0.0, regex=True)

gas.us = gas.us.astype(float)
diesel.us = diesel.us.astype(float)
gprod.oil = gprod.oil.astype(float)

dates = gas.date
price = gas.us/max(gas.us)
gprice = pd.DataFrame({'date' : dates, 'price' : price})

dates = diesel.date
price = diesel.us/max(diesel.us)
dprice = pd.DataFrame({'date' : dates, 'price' : price})

gprod.oil = gprod.oil/max(gprod.oil)
```

```python
start = min(min(gprice.date), min(dprice.date), min(gprod.date))
end = max(max(gprice.date), max(dprice.date), max(gprod.date))
dates = pd.date_range(start, end)
dates = pd.DataFrame(pd.Series(dates), columns=['date']).set_index('date')
```

```python
gprod.set_index('date', inplace=True)
dprice.set_index('date', inplace=True)
gprice.set_index('date', inplace=True)
```

```python
gmerged = dates.join(gprod, rsuffix='_y', how='outer')
gmerged = gmerged.join(gprice, rsuffix='_x', how='outer')
gmerged = gmerged.fillna(0)
gmerged = gmerged.reset_index()

dmerged = dates.join(gprod, rsuffix='_y', how='outer')
dmerged = dmerged.join(dprice, rsuffix='_x', how='outer')
dmerged = dmerged.fillna(0)
dmerged = dmerged.reset_index()
```

```python
pdates = gmerged.date.astype(datetime.datetime).values
```

```python
plt.figure(figsize=(10, 8))
ax = plt.subplot(111)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
plt.ylim(0.4, 1)
plt.xlim(min(pdates), max(pdates))
plt.yticks(np.arange(0, 1, 0.1), [str(x*100) + "%" for x in np.arange(0, 1, 0.1)], fontsize=14)
plt.xticks(fontsize=14)
plt.tick_params(axis="both", which="both", bottom="off", top="off",
                labelbottom="on", left="off", right="off", labelleft="on")
plt.plot(gmerged.date.astype(datetime.datetime).values,
         gmerged.oil.values, color='b')
plt.plot(gmerged.date.astype(datetime.datetime).values,
         gmerged.price.values, color='r')
plt.title('Average Domestic Gas Price and Oil Production', fontsize=17, ha="center")
plt.savefig("us_gas.png", bbox_inches="tight")  
```

```python
import seaborn as sns

# Compute the correlation matrix
data = gmerged[(gmerged.date > datetime.date(1990, 1, 1)) & (gmerged.date < datetime.date(2005, 1, 1))]
data = data.set_index('date')
corr = data.corr()

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=0.5,
            linewidths=.5, cbar_kws={"shrink": .5},
            annot=True, ax=ax)
```
